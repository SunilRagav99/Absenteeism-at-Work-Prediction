```python
# from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

ordinalencoder=OrdinalEncoder()
dff[['Gender',"City",'JobTitle',"DepartmentName",'StoreLocation',"Division","BusinessUnit"]]=ordinalencoder.fit_transform(dff[['Gender',"City",'JobTitle',"DepartmentName",'StoreLocation',"Division","BusinessUnit"]])
```

```python
dff
```

```python
dff["log_absent"]=np.log(dff["AbsentHours"])
```

```python
dff.loc[np.isneginf(dff['log_absent']), 'log_absent'] = 0
```

```python
X=dff[["Age","Gender","City","JobTitle","DepartmentName","StoreLocation","Division","LengthService","BusinessUnit"]]
```

```python
y=dff["log_absent"]
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
standcondition=StandardScaler()
X=standcondition.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# regressor
from sklearn.linear_model import LinearRegression
LinearR=LinearRegression()
LinearR.fit(X_train,y_train)
print(LinearR.score(X_train,y_train))
print(LinearR.score(X_test,y_test))
y_predict=LinearR.predict(X_test)
```

```python
# Calculate the Mean Squared Error
mse = mean_squared_error(y_test, y_predict)
print(f'Mean Squared Error: {mse}')

#Calculate the root Mean Squared Error
rmse=np.sqrt(mse)
print(f"RMSE:{rmse}")

# Calculate the R-squared score
r2 = r2_score(y_test, y_predict)
print(f'R-squared: {r2}')


# Print the model coefficients
# print("Intercept:", LinearR.intercept_)
# print("Coefficients:", LinearR.coef_)
```

```python
from sklearn.decomposition import PCA
PCAMLR=LinearRegression()
pca = PCA(n_components=9)  
X_pca = pca.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.4, random_state=42)

# Create and fit MLR PCA model
PCAMLR = LinearRegression()
PCAMLR.fit(X_train, y_train)

# Make predictions
y_pred = PCAMLR.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)

r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
#Calculate the root Mean Squared Error
rmse=np.sqrt(mse)
print(f"RMSE:{rmse}")
print(f'R-squared: {r2}')
# Print the model coefficients
# print("Intercept:", PCAMLR.intercept_)
# print("Coefficients:", PCAMLR.coef_)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Define the RandomForestRegressor
RFR = RandomForestRegressor()

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=RFR, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Print the best parameters
print("Best parameters found: ", grid_search.best_params_)

# Predict using the best estimator
y_pred_RFR = grid_search.best_estimator_.predict(X_test)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred_RFR)
r2 = r2_score(y_test, y_pred_RFR)
rmse = np.sqrt(mse)

# Print results
print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared: {r2}')

# Note: Exponentiation of predictions is typically not required for regression tasks.
# print(np.exp(y_pred_RFR))  # This line is commented out because it's usually not needed.
```
